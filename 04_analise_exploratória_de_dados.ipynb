OBSERVAÇÃO!! embora não seja incorreto substituir valores que parecem estar escritos de forma errada por categorias mais claras ou padronizadas, é importante lembrar que essa decisão é arbitrária. Ao fazer a substituição, estamos assumindo qual deveria ser o valor correto, mas não temos garantia de que essa interpretação está certa.

Por exemplo, ao trocar "MV" por "Movie" ou "Show" por "TV Show", estamos inferindo o significado com base no contexto do dataset. Em muitos casos isso faz sentido, mas ainda assim é uma suposição feita pelo analista.

Por isso, em situações reais de trabalho, o ideal é confirmar essas correções com a pessoa responsável pela base de dados ou com quem coletou as informações. Assim evitamos alterar dados de forma incorreta e garantimos que a padronização realmente representa o significado original dos registros.

## Converter Colunas

No dataset da Netflix, a coluna date_added representa a data em que o conteúdo foi adicionado à plataforma. No entanto, ao carregar o CSV, essa coluna geralmente é interpretada pelo pandas como texto (string).

Para poder trabalhar corretamente com datas — como extrair ano, mês ou fazer análises temporais — precisamos converter essa coluna para o tipo datetime.

Para isso utilizamos a função `pd.to_datetime()`.

## Criando uma nova variável
Agora que a coluna está no formato correto de data, podemos extrair partes específicas dessa informação.
Uma análise comum em datasets é observar em quais anos ocorreram mais eventos, como por exemplo quantos filmes foram adicionados à plataforma em cada ano.

Para isso, podemos criar uma nova coluna contendo apenas o ano em que o título foi adicionado e não a data completa.

```python
# criando uma nova coluna com o ano em que o conteúdo foi adicionado
df["year_added"] = df["date_added"].dt.year
df
```

Essa nova variável pode ser muito útil em análises futuras, por exemplo:

- identificar em quais anos mais conteúdos foram adicionados

- analisar o crescimento do catálogo ao longo do tempo

- comparar a quantidade de filmes e séries adicionados por ano

# Visualização de Dados
A visualização de dados é uma etapa fundamental da análise exploratória de dados. Ela consiste em representar informações por meio de gráficos e elementos visuais, permitindo compreender padrões, tendências e relações presentes nos dados de forma mais intuitiva.

Enquanto tabelas e números podem ser difíceis de interpretar rapidamente, gráficos ajudam a transformar grandes volumes de dados em representações visuais mais fáceis de entender. Isso permite identificar comportamentos que muitas vezes não seriam percebidos apenas observando os valores brutos.

```python
# quantidade de Filmes vs Séries
df["type"].value_counts()  # dado bruto
df["type"].value_counts().reset_index()
```

```python
type_counts = df["type"].value_counts().reset_index()

fig = px.bar(
    type_counts,
    x="type",
    y="count",
    labels={
        "type": "Tipo de conteúdo",
        "count": "Quantidade"
    },
    title="Quantidade de Filmes e Séries no Dataset"
)

fig.show() # grafico bonito e colorido
```

```python
conteudo_por_ano = df["year_added"].value_counts().sort_index()

fig = px.line(
    conteudo_por_ano,
    labels={"value": "Quantidade", "year_added": "Ano"},
    title="Quantidade de conteúdos adicionados por ano"
)

fig.show()
```

```python
top_paises = df["country"].value_counts().head(10)

fig = px.bar(
    top_paises,
    labels={"value": "Quantidade", "country": "País"},
    title="Países com mais conteúdos no dataset"
)

fig.show()
```
