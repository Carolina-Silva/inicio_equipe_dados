####  O "printf" do Python: Formatação de Textos (f-strings)

Como o Python não tem um comando `printf()`, nós alcançamos o mesmo resultado usando as **f-strings** (format strings). Elas foram introduzidas nas versões mais recentes do Python e são o padrão atual da indústria para formatar textos.

Para usar uma f-string, basta colocar a letra `f` minúscula ou maiúscula imediatamente antes das aspas do seu texto. Depois, você coloca qualquer variável ou cálculo dentro de chaves `{}` no meio da frase.

**Vantagens das f-strings:**
* **Legibilidade:** Você lê a frase exatamente como ela vai aparecer, sem precisar ficar somando pedaços de texto.
* **Formatação de Números:** É muito fácil arredondar casas decimais diretamente dentro das chaves, o que é extremamente útil quando estamos lidando com resultados estatísticos ou de modelos de dados.

```python
x = 3.14159

print(f"Pi é o numero: {x}")
print(f"Numero formatado com duas casas decimais: {x:.2f}")
print(f"Numero formatado com quatro casas decimais: {x:.4f} !")
```

## Variáveis

Variáveis guardam informações.  Pense nelas como "caixas" com etiquetas na memória do computador. Você coloca um dado lá dentro e usa a etiqueta (o nome da variável) toda vez que precisar daquela informação de novo ao longo do seu código.

No Python, você não precisa declarar qual é o tipo de dado que a variável vai guardar antes de criá-la; a linguagem descobre isso automaticamente no momento em que você atribui um valor a ela. 

```python
nome = "João da Silva"
idade = 19

print(nome)
print(idade)
```

Os quatro tipos básicos (primitivos) que você mais vai utilizar para estruturar e analisar dados são:

* **`int` (Inteiro):** Números inteiros, positivos ou negativos, sem casas decimais. Útil para contagens, como número de registros ou idades.
* **`float` (Ponto Flutuante):** Números com casas decimais. No Python, usamos o ponto (`.`) em vez da vírgula para separar as casas. Essencial para métricas, taxas e cálculos estatísticos.
* **`str` (String):** Textos, palavras ou caracteres. Devem ser sempre declarados entre aspas (simples `' '` ou duplas `" "`). Usado para categorias, nomes, ou códigos textuais.
* **`bool` (Booleano):** Só pode receber dois valores: Verdadeiro (`True`) ou Falso (`False`). Fundamental para aplicar filtros em tabelas e criar condições lógicas.

```python
numero = 10
decimal = 3.14
texto = "Python"
ativo = True
```

##  Convertendo tipos
Como vimos antes, o `input()` sempre entrega a informação no formato de texto (`str`). Isso é um cenário muito comum no dia a dia da análise de dados: ao importar bases ou ler arquivos externos, muitas vezes os números chegam ao Python disfarçados de texto.

Para podermos realizar cálculos matemáticos ou aplicar modelos estatísticos, precisamos converter explicitamente esses dados para formatos numéricos. Usamos comandos como `int()` (para números inteiros) ou `float()` (para decimais) envolvendo a informação que queremos transformar.

```python
idade = input("Digite sua idade: ")

idade = int(idade) # aqui idade recebe ele mesmo porem com a função int() convertendo o tipo

ano_futuro = idade + 10

print("Daqui 10 anos você terá", ano_futuro)
```

## Condições (IF /  ELSE)

Agora o Python começa a "pensar" e tomar decisões com base nos dados que recebe.

* **Lógica Condicional:** O uso do `if` (se) e `else` (senão) permite que o seu código crie bifurcações. Dependendo do valor de uma variável, o programa escolhe executar um bloco de código e ignorar o outro. Isso é a base para filtrar dados e criar regras de negócio.


```python
idade = int(input("Idade: "))

if idade >= 18:
    print("Você é maior de idade")
else:
    print("Você é menor de idade")
```

> * **Indentação (MUITO importante):** Diferente de outras linguagens que usam chaves `{}` para agrupar blocos de código, o Python usa o espaço em branco no início da linha (geralmente 4 espaços ou 1 *Tab*). A indentação não é apenas para deixar o código bonito; ela é obrigatória. É assim que o Python sabe exatamente quais comandos estão dentro do `if` e quais estão dentro do `else`. Se o alinhamento estiver errado, o programa vai gerar um erro.

### Mais condições

```python
nota = 7

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

## Listas

Até agora, nossas variáveis guardavam apenas um único valor por vez. Mas e se precisarmos armazenar uma coleção inteira de informações, como uma coluna inteira de dados? É aí que entram as Listas. 
No Python, criamos uma lista colocando os itens entre colchetes `[]`, separados por vírgula. 



Um detalhe muito importante: na programação, a contagem das posições (chamada de "índice") sempre começa no **zero**. Ou seja, o primeiro item da lista está na posição `0`, o segundo na posição `1`, e assim por diante. 

As listas também são dinâmicas. Você pode adicionar novos elementos ao final delas a qualquer momento usando a função `.append()`.

```python
frutas: list[str] = ["maçã", "banana", "uva"]

print(frutas)

print(frutas[0])

print(frutas[2])

frutas.append("laranja")

print(frutas)
```

Além de adicionar itens com o `.append()`, as listas possuem várias outras funcionalidades embutidas (chamadas de métodos) que facilitam muito a nossa vida na hora de limpar e organizar dados:

* **`len()`:** (De *length*, comprimento). Conta quantos itens existem dentro da lista. Essencial para saber o tamanho de uma base de dados.
* **`.remove()`:** Procura por um valor específico que você digitar e apaga a primeira aparição dele na lista.
* **`.sort()`:** Organiza a lista. Se for texto, coloca em ordem alfabética. Se for número, coloca em ordem crescente.

```python
frutas = ["uva", "maçã", "banana", "laranja"]
print(frutas)

quantidade = len(frutas)
print(quantidade)

frutas.remove("maçã")
print(frutas)

frutas.sort()
print(frutas)
```

## Loops (repetição)

Agora entramos no verdadeiro poder da programação: a **automação**.


Imagine que você tenha uma lista com milhares de registros de um banco de dados de saúde e precise aplicar um cálculo para cada linha. Em vez de escrever o mesmo comando milhares de vezes, usamos um **Loop** (laço de repetição).
O loop `for` pega uma coleção (como a nossa lista de frutas), olha para o primeiro item, executa os comandos que estão indentados logo abaixo, e depois volta para fazer a mesma coisa com o segundo item, repetindo o processo até a lista acabar.
Além de listas, muitas vezes queremos apenas repetir uma ação um número específico de vezes. Para isso usamos o `range()`. Como o Python sempre começa a contar do zero, o comando `range(5)` vai gerar os números 0, 1, 2, 3 e 4.

```python
for fruta in frutas:
    print(fruta)

```

```python
for i in range(5): # Loop numérico
    print(i)
```

##  Funções (Criando seus próprios comandos)
Até agora, usamos funções que já vêm prontas no Python, como o `print()`, `input()` e `len()`. Mas para organizar análises maiores e não ficar copiando e colando o mesmo código várias vezes, nós criamos as nossas próprias funções.

Pense em uma função como uma mini-fábrica dentro do seu código: você entrega os materiais (chamados de **parâmetros**), ela faz o processamento lá dentro, e te devolve um produto pronto (usando o comando **`return`**). 

Para criar uma função, usamos a palavra `def` (de definir). Você escreve a lógica apenas uma vez e depois pode "chamar" essa função quantas vezes precisar!

```python
def dar_boas_vindas(nome: str) -> str:
    mensagem = f"Olá, {nome}! Parabéns por concluir esta revisão."
    return mensagem

texto_saudacao = dar_boas_vindas("aluno")
print(texto_saudacao)

```

```python
def calcular_dobro(numero: float) -> float:
    return numero * 2

resultado_dobro = calcular_dobro(15.5)
print(resultado_dobro)
```
